```python
import warnings                                 
warnings.filterwarnings('ignore')

import numpy as np                               
import pandas as pd                              
import matplotlib.pyplot as plt                 
import seaborn as sns
```

```python
ads_df = pd.read_csv('ads.csv')
currency_df = pd.read_csv('currency.csv')
```

```python
ads_df
```

```python
currency_df
```

```python
ads_df['Time'] = pd.to_datetime(ads_df['Time'])
currency_df['Time'] = pd.to_datetime(currency_df['Time'])
ads_df.set_index('Time',inplace=True)
currency_df.set_index('Time',inplace=True)
```

```python
plt.figure(figsize=(15,7))
plt.plot(currency_df['GEMS_GEMS_SPENT'])
plt.title('Daily GEMS_GEMS_SPENT')
plt.grid(True)
```

```python
plt.figure(figsize=(15,7))
plt.plot(ads_df.Ads)
plt.title('Hourly Ads Data')
plt.grid(True)
```
